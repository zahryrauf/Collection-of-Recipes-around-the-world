**Collection of Recipes around the world**

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("prajwaldongre/collection-of-recipes-around-the-world")

print("Path to dataset files:", path)
```

**View Dataset**

```python
import pandas as pd

# Try reading the CSV file with 'latin-1' encoding
df = pd.read_csv(path + "/Receipes from around the world.csv", encoding='latin-1')
# If 'latin-1' doesn't work, try other encodings like 'ISO-8859-1', 'cp1252', etc.

df
```

**Basic cleaning**

```python
df = df.drop_duplicates()  # Remove duplicates
df = df.dropna()            # Remove rows with missing values
```

**Calculate the cosine similarity between ingredient vectors to predict the most similar recipe** **bold text**

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Assuming 'df' is your DataFrame with an 'ingredients' column
# Example:
# df = pd.DataFrame({'ingredients': ['flour, sugar, eggs', 'sugar, butter, flour', 'milk, sugar']})

# Create a TF-IDF vectorizer
tfidf = TfidfVectorizer()

# Fit and transform the ingredients column
tfidf_matrix = tfidf.fit_transform(df['ingredients'])

# Calculate cosine similarity between all recipe pairs
cosine_similarities = cosine_similarity(tfidf_matrix)

# Example usage: find recipes most similar to the first recipe
recipe_index = 0  # Index of the first recipe
similar_recipes = list(enumerate(cosine_similarities[recipe_index]))

# Sort the recipes based on similarity scores (excluding the recipe itself)
sorted_similar_recipes = sorted(similar_recipes[1:], key=lambda x: x[1], reverse=True)

# Print the top N similar recipes
N = 5 # you can change the value here
print(f"Top {N} similar recipes to recipe {recipe_index}:")
for i, score in sorted_similar_recipes[:N]:
    print(f"Recipe index {i}, Similarity: {score}")
    print(f"Ingredients: {df['ingredients'][i]}")
```
